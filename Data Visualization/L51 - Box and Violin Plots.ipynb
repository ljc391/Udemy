```python
import numpy as np
from numpy.random import randn
import pandas as pd

#Import the stats library from numpy
from scipy import stats

#These are the plotting modules libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the jupyter notebook
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1, data2])
```

```python
sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
#set orient v/h
sns.boxplot(data = data1, orient="v")
```

```python
#Normal Dist
data1 = stats.norm(0,5).rvs(100)

#Two gammna dist. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, 
                       stats.gamma(5).rvs(50)])

#Box plot both data1 and data2
sns.boxplot(data =[data1, data2], whis=np.inf)
```

```python
sns.violinplot(data = [data1, data2])
```

```python
#breadth
sns.violinplot(data = data2, bw = 0.01)
```
